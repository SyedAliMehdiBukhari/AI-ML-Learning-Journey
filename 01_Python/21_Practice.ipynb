### Practice Q2

Calculate Average Age of 10 Students

```python
i = 1
students_age = []
total_age = 0

while (i < 11):
    student = int(input(f"Enter Age of Student {i}: "))
    students_age.append(student)
    i+= 1

for age in students_age:
    total_age += age
    average = total_age/len(students_age)

print(f"Average age of {len(students_age)} Students is: {average}")
```

### Practice Q3

Take 4-digits input from user and show it in reverse order with digits separated by comma

```python
four_digit_number = int(input(f"Enter a 4-digit number: "))

for i in range(4):
    last_digit = four_digit_number % 10
    print(last_digit, end=", ")
    four_digit_number = four_digit_number // 10
```

1 more method

```python
number_4_digit = input("Enter a 4-digit number: ")

for number in number_4_digit[::-1]:
    print(number, end=", ")
```

### Practice Q4 

Program which takes radius of a cricle from user and gives it's diameter, circumference and area of circle

```python
import math

radius = float(input("Enter radius of a circle: "))

diameter = 2 * radius
circumference = 2 * math.pi * radius
area = math.pi * (radius ** 2)

print(f"The Diameter of the circle is {diameter}, circumference of the circle is {circumference:.2f} and area of circle is {area:.2f}")
```

### Practice Q6

Program that calculates the Body Mass Index (BMI) from a user's weight and height.

```python
weight = float(input("Enter the weight of the person in (Kg): "))
height = float(input("Enter the height of the person in (m): "))

bmi = weight / (height ** 2)

print(f"The (BMI) Body Mass Index of person is: {bmi:.2f}")
```

### Practice Q7

Coffee Shop

```python
PRICE_PER_CAPPUCCINO = 4.5
TAX_RATE = 0.08
ELITE_DISCOUNT_RATE = 0.05

quantity = 3
is_elite_member = True

sub_total = PRICE_PER_CAPPUCCINO * quantity

discount_rate = ELITE_DISCOUNT_RATE if is_elite_member else 0.0

discount_amount = sub_total * discount_rate
discounted_price = sub_total - discount_amount

total_tax = discounted_price * TAX_RATE

final_bill = discounted_price + total_tax

print(f"Subtotal: ${sub_total:.2f}")
print(f"Discount: -${discount_amount:.2f}")
print(f"Tax: +${total_tax:.2f}")
print(f"Final Bill: ${final_bill:.2f}")
```
